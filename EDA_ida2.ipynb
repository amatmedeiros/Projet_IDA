*a) pour un même produit et pour une même semaine, remplissage par les valeurs présentes dans df_ida1*

**Création du catalogue ida2**

```python
df_ida2_catalog = df_ida2[['année', 'nom_produit', 'semaine_iso', 'prix_vente', 'colisage']].drop_duplicates()
df_ida2_catalog.sort_values(by=(['nom_produit', 'semaine_iso']), inplace=True)
df_ida2_catalog.reset_index(drop=True, inplace=True)
df_ida2_catalog.info()
```

***Gestion de prix de ventes différents pour un même produit, dans une même semaine d'une même année***

*Vérification de doublons*

```python
doublons = df_ida2_catalog[df_ida2_catalog.duplicated(subset=['nom_produit', 'année', 'semaine_iso'], keep=False)]
doublons
```

*Fonction pour consolider les prix_vente :*

1) Consolidation du prix_vente le plus fréquent (mode) ;

2) Si deux prix_vente, consolidation de celui plus proche du début de la semaine.

```python
def prix_consolide_groupe(group):
    prix_counts = group['prix_vente'].value_counts(dropna=True)
    if prix_counts.empty:
        return np.nan  # Pas de prix disponible dans le groupe

        #dropna=True: ignore les NaN. Si cette option n'était pas mise, les NaN seraient inclus dans le comptage comme une "valeur".

    max_freq = prix_counts.max()
    candidats = prix_counts[prix_counts == max_freq].index.tolist()

    if len(candidats) == 1:
        return candidats[0]  # Prix le plus fréquent
    else:
        # En cas d'égalité, on prend celui de la date la plus ancienne (début semaine)
        subset = group[group['prix_vente'].isin(candidats)].copy()
        if subset.empty:
            return np.nan

        # Calculer date du lundi pour cette semaine/année
        year = group['année'].iloc[0]
        week = group['semaine_iso'].iloc[0]
        date_lundi = pd.to_datetime(f'{year}-W{int(week):02d}-1', format='%G-W%V-%u')

        subset['distance_lundi'] = (subset['date_vente'] - date_lundi).abs()
        closest_idx = subset['distance_lundi'].idxmin()
        return subset.loc[closest_idx, 'prix_vente']

df_ida2_catalog = (
    df_ida2.groupby(['nom_produit', 'année', 'semaine_iso'])
    .apply(prix_consolide_groupe)
    .reset_index(name='prix_vente')
)
```

*Consolidation du colisage : le plus fréquent (mode)*

```python
colisage_mode = (
    df_ida2
    .groupby(['nom_produit', 'année', 'semaine_iso'])['colisage']
    .agg(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
    .reset_index(name='colisage')
)

# Fusionner les deux pour avoir le catalogue complet
df_ida2_catalog = pd.merge(df_ida2_catalog, colisage_mode, on=['nom_produit', 'année', 'semaine_iso'])
```

**Merge des deux catalogues**

```python
#Renommage des colonnes pour différencier les prix de ventes et les colisages de df_ida1_catalog et de df_ida2_catalog

df_ida1_catalog = df_ida1_catalog.rename(columns={
    'prix_vente': 'prix_vente_catalog_1',
    'colisage': 'colisage_catalog_1',
    'year': 'année'
})

df_ida2_catalog = df_ida2_catalog.rename(columns={
    'prix_vente': 'prix_vente_catalog_2',
    'colisage': 'colisage_catalog_2'
})
```

```python
#Inner join pour ne garder que les lignes où nous avons les mêmes produits, les mêmes semaines et les mêmes années pour df_ida1_catalog et de df_ida2_catalog

df_comparaison = pd.merge(
    df_ida1_catalog, df_ida2_catalog,
    on=['nom_produit', 'semaine_iso', 'année'],
    how='inner')
```

```python
#Remplissage des NaN des colonnes "_catalog_2" par les valeurs des colonnes "_catalog_1" (pour le même produit, même semaine_iso et même année)

df_comparaison['prix_vente_catalog_2'] = df_comparaison['prix_vente_catalog_2'].fillna(df_comparaison['prix_vente_catalog_1'])
df_comparaison['colisage_catalog_2'] = df_comparaison['colisage_catalog_2'].fillna(df_comparaison['colisage_catalog_1'])

df_comparaison
```

**Merge df_comparaison et df_ida2 pour récupérer les valeurs désormais remplies (prix_vente_catalog_2	colisage_catalog_2)**

```python
# Étape 1 : jointure gauche pour ajouter les colonnes du catalogue
df_temp = df_ida2.merge(
    df_comparaison[['nom_produit', 'année', 'semaine_iso', 'prix_vente_catalog_2', 'colisage_catalog_2']],
    on=['nom_produit', 'année', 'semaine_iso'],
    how='left'
)

# Étape 2 : remplacer les valeurs manquantes
df_temp['prix_vente'] = df_temp['prix_vente'].fillna(df_temp['prix_vente_catalog_2'])
df_temp['colisage'] = df_temp['colisage'].fillna(df_temp['colisage_catalog_2'])

# Étape 3 : supprimer les colonnes temporaires
df_ida2 = df_temp.drop(columns=['prix_vente_catalog_2', 'colisage_catalog_2'])

df_ida2
```

### *b) remplissage des valeurs nulles par les valeurs les plus proches dans le temps (pour un même produit)*

*Tri selon le nom du produit et la date de la vente*

```python
df_ida2 = df_ida2.sort_values(by=['nom_produit', 'date_vente'])
df_ida2
```

*Fonction pour remplir les valeurs nulles des colonnes concernées par les valeurs d'autres colonnes plus proche dans le temps (pour un même produit)*

*Calcul de la "qt_vendue" d'après les nouvelles valeurs de la colonne "prix_vente"*

```python
mask = (
    df_ida2['qt_vendue'].isna() &
    df_ida2['montant_vente'].notna() &
    df_ida2['prix_vente'].notna()
)

df_ida2.loc[mask, 'qt_vendue'] = (
    df_ida2.loc[mask, 'montant_vente'] / df_ida2.loc[mask, 'prix_vente']
)
```
