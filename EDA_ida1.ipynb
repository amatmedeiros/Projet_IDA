# **I. Nettoyage initial de la table df_ida1**

```python
## Import du fichier ida dans le DF df_ida1
df_ida1 = pd.read_csv(f'{file_dir}/Datasets/Fichiers_Source/df_ida.csv')
df_ida1
```

```python
print(df_ida1.info())

print(df_ida1.describe())
```

**1. Analyse et suppression des doublons**

```python
# Nombre de lignes en doublon avant suppression.
print(f"Nombre de lignes en doublon avant suppression : {df_ida1.duplicated().sum()}")
```

```python
# Suppression des lignes en doublon
df_ida1.drop_duplicates(inplace=True)

#Nombre de lignes en doublon aprÃ¨s suppression
df_ida1.reset_index(drop=True, inplace=True)
print(f"Nombre de lignes en doublon aprÃ¨s suppression : {df_ida1.duplicated().sum()}")
```

```python
print(df_ida1.info())

print(df_ida1.describe())
```

**2. Conversion des donnÃ©es et alimentation de la colonne PROMO**

```python
# Conversion de la date au format DATE
df_ida1['date'] = pd.to_datetime(df_ida1['date'])
df_ida1.info()
```

```python
# Alimentation des champs PROMO vide en 'Hors Promo'
df_ida1['PROMO'][df_ida1['PROMO'] == ' '] = 'Hors Promo'
df_ida1
```

**3. Suppression des colonnes inutiles**

```python
# Suppression des colonnes weekday, week
colonnes_a_supprimer = ["weekday", "week"]
df_ida1 = df_ida1.drop(columns=colonnes_a_supprimer)
df_ida1.info()
```

**4. Renaming des colonnes**

```python
df_ida1.rename(columns=
 {'store' : 'nom_magasin',
  'libelle' : 'nom_produit',
  'CA' : 'montant_vente',
  'date' : 'date_vente',
  'PV' : 'prix_vente',
  'PCB' : 'colisage',
  'sold_qty' : 'qt_vendue'}
 , inplace=True)
df_ida1.info()
```

**5. Nettoyage des noms de produits**

```python
## Remplace la double-cote par un espace
df_ida1['nom_produit'] = df_ida1['nom_produit'].str.replace(r'["]', ' ', regex=True)
## Remplace les caqractÃ¨res spÃ©ciaux par rien
df_ida1['nom_produit'] = df_ida1['nom_produit'].str.replace(r'[Â£#,%*]', ' ', regex=True)
df_ida1['nom_produit'] = df_ida1['nom_produit'].str.replace('AÂ£', ' ', regex=True)
## Remplace plusieurs espaces par un seul
df_ida1['nom_produit'] = df_ida1['nom_produit'].str.replace(r'\s+', ' ', regex=True)
## Suppression des espace en dÃ©but et fin de chaÃ®ne
df_ida1['nom_produit'] = df_ida1['nom_produit'].str.strip()
df_ida1['nom_produit'].unique()
```

# **II. Uniformisation de la colonne colisage**

```python
# VÃ©rification de l'existence de valeurs nulles ou aberrantes dans la colone 'colisage'
colisage_anomalie = df_ida1['colisage'].unique()
colisage_anomalie.sort()
print(colisage_anomalie)
```

```python
# Nombre de valeurs de colisage par produit
colisage_par_produit = df_ida1.groupby('nom_produit')['colisage'].nunique()
print(colisage_par_produit)
```

```python
# Identification des produits portant plus d'un colisage
anomalie_colisage = colisage_par_produit[colisage_par_produit > 1]
print(anomalie_colisage)
```

```python
# DÃ©tail des produits portant plus d'un colisage
colisage_sup_1 = df_ida1.groupby('nom_produit')['colisage'].unique()[colisage_par_produit > 1]
print(colisage_sup_1)
```

```python
# Calcul du nombre de vente par produit et par colisage
df_nb_ventes = df_ida1.groupby(['nom_produit', 'colisage']).size().reset_index(name='nb_ventes')
print(df_nb_ventes)
```

```python
# Conservation des colisages portant le plus grand nombre de ventes
df_nb_ventes_max = df_nb_ventes.loc[df_nb_ventes.groupby('nom_produit')['nb_ventes'].idxmax()]
print(df_nb_ventes_max)
```

```python
# Renaming de la colonne 'colisage' du df "df_nb_ventes_max"

df_nb_ventes_max.rename(columns={'colisage' : 'colisage_max'}, inplace=True)
print(df_nb_ventes_max)
```

```python
# Fusion des DataFrames df_ida1 et df_nb_ventes_max

df_ida1 = df_ida1.merge(df_nb_ventes_max, on=['nom_produit'], how='left')
df_ida1
```

```python
# Suppression des colonnes 'colisage' de df_ida1 et 'nb_ventes' de df_nb_ventes_max

df_ida1.drop(columns=(['colisage', 'nb_ventes']), inplace=True)

# Renaming de la colonne 'colisage_max' en 'colisage'

df_ida1.rename(columns={'colisage_max' : 'colisage'}, inplace=True)
df_ida1
```

```python
# Nombre de lignes en doublon avant suppression.
print(f"Nombre de lignes en doublon avant suppression : {df_ida1.duplicated().sum()}")
```

```python
# Suppression des doublons

df_ida1=df_ida1.drop_duplicates()
df_ida1.reset_index(drop=True, inplace=True)
df_ida1
```

```python
#Nombre de lignes en doublon aprÃ¨s suppression
print(f"Nombre de lignes en doublon aprÃ¨s suppression : {df_ida1.duplicated().sum()}")
```

# **III. Analyse et correction des ventes en doublons (prix de vente et quantitÃ© vendue diffÃ©rentes)**

```python
# Liste du nombre de valeurs de prix de vente par vente

df_prix_vente = df_ida1.groupby(['nom_magasin', 'nom_produit', 'montant_vente', 'date_vente'])['prix_vente'].nunique()
df_prix_vente
```

```python
# Liste des ventes portant plus d'une valeur de prix de vente

df_prix_vente[df_prix_vente > 1]
```

```python
# Classement par valeur de la colonne 'prix_vente'

df_ida1.sort_values(by=['prix_vente'], ascending=[True], inplace=True)

# Suppression des lignes en doublon portant pas la valeur de prix de vente maximale - pour ne pas supprimer des lignes portant la mention 'Promo'

df_ida1.drop_duplicates(subset=['nom_magasin',
                                'nom_produit',
                                'montant_vente',
                                'date_vente',
                                'colisage'], keep='first', inplace=True)
df_ida1.reset_index(drop=True, inplace=True)
df_ida1
```

```python
df_ida1.info()
```

# **IV. Analyse et correction de la colonne "PROMO"**

```python
# Nombre de lignes portant la mention 'Promo'

nb_lignes_promo = len(df_ida1[df_ida1['PROMO'] == 'Promo'])
print(f"Nombre de lignes avec la valeur 'Promo' : {nb_lignes_promo}")
```

```python
# Nombre de produits portant la mention 'Promo'

nb_produit_promo = df_ida1[df_ida1['PROMO'] == 'Promo']['nom_produit'].nunique()
print(f"Nombre de produits avec la valeur 'Promo' : {nb_produit_promo}")
```

```python
# Regroupement des lignes par nombre de valeurs dans la colonne 'PROMO'
promo_par_produit = df_ida1.groupby(['nom_magasin',
                                     'nom_produit',
                                     'montant_vente',
                                     'date_vente',
                                     'prix_vente',
                                     'colisage',
                                     'qt_vendue'])['PROMO'].nunique()
promo_par_produit
```

```python
# Identification des lignes en double, portant plus d'une valeur dans la colonne 'Promo'

promo_par_produit[promo_par_produit > 1]
```

```python
# Classement par valeur de la colonne 'PROMO'

df_ida1.sort_values(by=['PROMO'], ascending=[False], inplace=True)

# Suppression des lignes en doublon ne portant pas la valeur 'Promo'

df_ida1.drop_duplicates(subset=['nom_magasin',
                                'nom_produit',
                                'montant_vente',
                                'date_vente',
                                'prix_vente',
                                'colisage',
                                'qt_vendue'], keep='first', inplace=True)
df_ida1.reset_index(drop=True, inplace=True)
df_ida1
```

```python
print(df_ida1.info())
```

```python
# Liste des produits portant la mention 'Promo'

list_produit_promo = df_ida1[df_ida1['PROMO'] == 'Promo']['nom_produit'].unique()
list_produit_promo
```

```python
# CrÃ©ation d'un DF des produits portant au moins sur une vente la mention 'Promo'

df_produit_promo = df_ida1[df_ida1['nom_produit'].isin(list_produit_promo)]
df_produit_promo.reset_index(drop=True, inplace=True)
df_produit_promo
```
